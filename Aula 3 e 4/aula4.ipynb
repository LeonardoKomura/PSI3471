```python
import cv2
import queue
import numpy as np
import math 
```

```python
def bucket(a, i, j):
    print(i, j)
    #print(pixel)
    #print(math.dist(pixel, a[i-1,j]))
    if(math.dist(a[i,j], a[i-1,j]) < 50):
        a[i,j] = [0, 0, 255]
        bucket(a, i-1, j)
    if(math.dist(a[i,j], a[i,j-1]) < 50):
        a[i,j] = [0, 0, 255]
        bucket(a, i, j-1)
    if(math.dist(a[i,j], a[i+1,j]) < 50):
        a[i,j] = [0, 0, 255]
        bucket(a, i+1, j)
    if(math.dist(a[i,j], a[i,j+1]) < 50):
        a[i,j] = [0, 0, 255]
        bucket(a, i, j+1)
```

```python
a = cv2.imread('elefante.jpg',cv2.IMREAD_COLOR)
```

```python
bucket(a, 130, 130)
```
