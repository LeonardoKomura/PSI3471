```python
import cv2
import queue
import numpy as np
import math 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#1 = (140,45) -> (115,10)
#2 = (322, 34) -> (345,10)
#3 = (108,295) -> (115,295)
#4 = (351, 294) -> (345, 295)
```

```python
src = np.array([[140, 45], [322, 34], [108, 295], [351, 294]], np.float32)
print(src)
```

```python
dst = np.array([[115, 10], [345, 10], [115, 295], [345, 295]], np.float32)
print(dst)
```

```python
m = cv2.getPerspectiveTransform(src, dst)
print(m)
```

```python
calib = cv2.imread('calib_result.jpg',cv2.IMREAD_COLOR)
img = mpimg.imread('calib_result.jpg')
imgplot = plt.imshow(img)
plt.show()
```

```python
out =  cv2.warpPerspective(calib, m, (len(calib[0]), len(calib)))
cv2.imwrite('tabuleiro.png', out)
img = mpimg.imread('tabuleiro.png')
imgplot = plt.imshow(img)
plt.show()
```
